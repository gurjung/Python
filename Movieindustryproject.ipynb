```python
# importing library files
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjust the configuation of the plots we create
```

```python
df.dropna(inplace=True)
```

```python
round(np.mean(df.isnull())*100)
```

```python
df.isnull()
```

```python
# Checking and Changing data types if required
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
df['score'] = df['score'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.dtypes
```

```python
# Looking at the top 15 compaies by gross revenue
df.groupby('company')[["gross"]].sum().sort_values('gross',ascending=False)[:15]
```

```python
# Scatter plot between Budget and Gross
sns.scatterplot(data=df, x="budget", y="gross").set(title = 'Budget vs Gross Earnings')
```

```python
# regression plot for budget vs gross earnings

sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
# rating vs gross strip plot
sns.stripplot(data=df,x='rating',y='gross')
plt.title('Rating vs Gross Earnings')
```

```python
#correlation matrix
df.corr() 
```

```python
# heatmap of correlation matrix
corr_matrix = df.corr()
sns.heatmap(corr_matrix,annot=True)
plt.title('Correlation matrix for numeric features')
plt.xlabel('Movies Features')
plt.ylabel('Movies Features')
plt.show()
```
